# Generate Sinasoid

## Real Sinasoid

$$
x[n] = A \cos(2 \pi f nT + \phi)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

A = 0.8
f0 = 1000
phi = np.pi / 2
fs = 44100
t = np.arange(-0.002,0.002,1.0/fs)
x = A * np.cos( 2 * np.pi * f0 * t + phi)

plt.plot(t,x)
```

## Complex Sinasoid

$$
S_k[n] = e^j 2\pi kn /N = \cos(2 \pi kn / N ) + j \sin(2 \pi kn / N)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 500
k = 3
n = np.arange(-N/2, N/2)

s = np.exp(1j  * 2 * np.pi *k *n / N)

plt.plot(n,np.real(s))
```
